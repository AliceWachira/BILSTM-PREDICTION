### **LOADING DATA**

The data below contains 12 csv files for each month. All the files are concatenated to one dataframe.

```python
import os, glob


path = "/content/gdrive/MyDrive/AIMS ESSAY/TEST DATASET"

all_files = glob.glob(os.path.join(path, "*.csv"))
df_from_each_file = (pd.read_csv(f, sep=',') for f in all_files)
df_merged   = pd.concat(df_from_each_file, ignore_index=True)
#df_merged.to_csv( "merged.csv")
```

```python
df_merged
```

Deleting the irrelevant columns 6 and 7.

```python
del df_merged['Unnamed: 6']
del df_merged['Unnamed: 7']
del df_merged['Status']
```

```python
del df_merged['Id'] #Deletes the initial indexing of rows and now creates a new Id
df_merged['Id']= df_merged.index + 1

```

```python
#FUNCTION TO SPLIT THE DataAsOf column and renaming the column

def col_splitter(df_merged, orig_col, deli):
    #Split the column on the delimiter provided
    df_split=df_merged[orig_col].str.split(deli, expand=True)
    
    # make a dictionary of new column names based on the number of splits
    col_names={k:f'{orig_col}_{k}' for k in  df_split.columns.values}
    
    # Lastly, rename columns 
    df_split.rename(columns=col_names,inplace=True)
    return  df_split



col_splitter(df_merged,'DataAsOf',' ')
```

```python
df_split=col_splitter(df_merged,'DataAsOf',' ')
#Crete an Id for the rows
df_split['Id']= df_split.index + 1
df_split
```

```python
 #Here we merge both the splitted dataframe containing the date and time and also the initial dataframe
 
 final_merge=pd.merge(df_merged, df_split, on='Id', how='outer')
  
 #Delete DateAsOf column
 del final_merge['DataAsOf']
 final_merge
```

```python
#Extract all the unique linkId in the dataset
final_merge['linkId'].unique()
```

```python
#Rename the Date and Time column

final_merge.columns=['Speed','TravelTime','linkId','Id','Date','Time']
final_merge
```

```python
#Convert the date into the datetime standard format
final_merge['Date']=pd.to_datetime(final_merge['Date'])

#Sort the data by date and time
final_sort=final_merge.sort_values(by=['Date','Time'])
final_sort
```

## **PATH SUB-SAMPLING**
The data was taken from a database of traffic information in New York City. The research region is a road network of 138 road pathways, with average speed and travel time for each path recorded from December 31, 2015 to December 31, 2016, with a 5-minute time interval.
To save on the training time, a sample of paths is considered in this study. In this reseach project, we worked with a sample of 28 paths out of the 138 paths in the dataset to decrease the sample size and speed up the training of the model. The criteria used to choose the paths was influenced by the number of boroughs considered in the collection of the NYC traffic data. The traffic data was recorded for five boroughs namely: Bronx, Brooklyn, Manhattan, Queens and Staten Island. A fifth of the paths in the five boroughs were considered.

```python
#Defines a function that extracts dataframes with respect to the linkId and saves them in the drive

def create_link_data(final_sort):
  for link in final_sort['linkId'].unique():
    df= final_sort.loc[final_sort['linkId']==link]
    #df=df.sort_values(by=['Date','Time'])# Sort by Date and Time
    df.to_csv('link_num'+str(link)+'.csv',index=False)
  return 
    
```
